## Part 2 - 가설 검정

### Hypothesis Test    
   - 수치형 (numerical/continuous) 칼럼이 어떤 차이를 보이는지 t-test를 통해 가설을 세우고 검증하세요. (최소 2개)
   - 범주형 (categorical) 칼럼이 어떤 차이를 보이는지 $\chi^2-test$를 통해 가설을 세우고 검증하세요. (최소 2개)  
  **Hint:** categorical feature 간의 "cross tab"을 생성하여 탐색해 보세요 (aka [contingency tables](https://en.wikipedia.org/wiki/Contingency_table)). 그리고 $\chi^2 test$를 적용해 보세요. [pandas.crosstab](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html)을 사용하면 contingency table 을 생성할 수 있습니다. 그리고 [scipy.stats.chi2_contingency](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html)를 사용하면 $\chi^2$ statistics 를 계산할 수 있습니다.    

```python
# 답안 작성은 이 곳에 하십시오
from scipy import stats
from scipy.stats import ttest_1samp
from scipy.stats import ttest_ind, chi2_contingency
```

```python
#귀무가설: 소비자들의 고기와 생선 구매량에는 차이가 있을 것이다
#대립가설: 소비자들의 고기와 생선 구매량에는 차이가 없을 것이다
#신뢰도: 95%
ttest_ind(df['MntMeatProducts'], df['MntFishProducts'])
```


첫번째 ttest pvalue가 0.05보다 크게 나왔으므로 대립가설을 채택한다



고기와 생선의 판매량에는 유의미한 차이가 있다.




```python
#귀무가설: 소비자들의 평균 출생년도는 1980년도 일 것이다.
#대립가설: 소비자들의 평균 출생년도는 1980년도가 아닐 것이다.
#신뢰도: 95%
agem = pd.DataFrame(df, columns = ['Year_Birth'])
stats.ttest_1samp(agem, 1980)
```



TTEST 결과 P-VALUE가 0.05보다 크므로 대립가설을 채택한다


즉 소비자들의 평균 출생년도는 1980년이 아니다

```python
#귀무가설: 소비자들의 평균 출생년도는 1980년도 일 것이다.
#대립가설: 소비자들의 평균 출생년도는 1980년도가 아닐 것이다.
#신뢰도: 95%
chi2_contingency(pd.crosstab(df['Education'], df['Country']))
```

```python

#귀무가설: 국적과 컴플레임 여부는 관련이 있을 것이다
#대립가설: 국적과 컴플레임 여부는 관련이 없을 것이다
#신뢰도: 95%
# p value가 0.05를 넘으므로 컴플레임을 거는데에는 국적이 관계가 없다는 것을 알 수 있다

stats.chisquare(pd.crosstab(df['Complain'], df['Country']))
```

```python

#귀무가설: 나이와 학력에는 연관관계가 있을 것이다.
#대립가설: 나이와 학력에는 연관관계가 없을 것이다.
#신뢰도: 95%
# p value가 0.05를 넘으므로 나이와 학력에는 연관관계가 없다.
stats.chisquare(pd.crosstab(df['Year_Birth'], df['Education']))
```
