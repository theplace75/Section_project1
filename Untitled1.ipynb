```python
import pandas as pd
import numpy as np
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv'

ktng_url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv'
#df = pd.read_csv(ktng_url, names = headers)
df = pd.read_csv(url, header = None)
```

```python
df_columns = df.iloc[0] #columns 값 저장

df.columns=df_columns #컬럼값 지정

df = df[1:]  # 표 정리

df
```
